This function is to plot the confusion matrix.


```python
def plot_confusion_matrix(y,y_predict):
    "this function plots the confusion matrix"
    from sklearn.metrics import confusion_matrix

    cm = confusion_matrix(y, y_predict)
    ax= plt.subplot()
    sns.heatmap(cm, annot=True, ax = ax); #annot=True to annotate cells
    ax.set_xlabel('Predicted labels')
    ax.set_ylabel('True labels')
    ax.set_title('Confusion Matrix'); 
    ax.xaxis.set_ticklabels(['did not land', 'land']); ax.yaxis.set_ticklabels(['did not land', 'landed']) 
    plt.show() 
```

```python
import numpy as np

# Predictions from each best estimator
y_lr  = logreg_cv.predict(X_test)
y_svm = svm_cv.predict(X_test)
y_dt  = tree_cv.predict(X_test)
y_knn = knn_cv.predict(X_test)

# Do any models produce identical predictions?
print("LR == SVM:", np.mean(y_lr  == y_svm))
print("LR == DT :", np.mean(y_lr  == y_dt))
print("SVM == DT:", np.mean(y_svm == y_dt))
print("KNN == DT:", np.mean(y_knn == y_dt))

# Class balance on the test set
import numpy as np
print("Test positive rate:", np.mean(Y_test))
```

```python
from collections import Counter
maj = Counter(Y_train).most_common(1)[0][0]
baseline_acc = np.mean(Y_test == maj)
print("Baseline (majority-class) accuracy:", baseline_acc)
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

# Pipelines
pipe_lr  = make_pipeline(StandardScaler(), LogisticRegression(max_iter=1000))
pipe_svm = make_pipeline(StandardScaler(), SVC())
pipe_knn = make_pipeline(StandardScaler(), KNeighborsClassifier())
pipe_dt  = DecisionTreeClassifier(random_state=42)  # trees don’t need scaling

# Grids
grid_lr  = {'logisticregression__C':[0.01,0.1,1], 'logisticregression__penalty':['l2'], 'logisticregression__solver':['lbfgs']}
grid_svm = {'svc__kernel':['linear','rbf','poly','sigmoid'], 'svc__C':np.logspace(-3,3,5), 'svc__gamma':np.logspace(-3,3,5)}
grid_knn = {'kneighborsclassifier__n_neighbors':range(1,11), 'kneighborsclassifier__algorithm':['auto','ball_tree','kd_tree','brute'], 'kneighborsclassifier__p':[1,2]}
grid_dt  = {'criterion':['gini','entropy'], 'splitter':['best','random'], 'max_depth':[2*n for n in range(1,10)],
            'max_features':[None, 'sqrt', 'log2'], 'min_samples_leaf':[1,2,4], 'min_samples_split':[2,5,10]}

# Grid searches (cv=10 as before)
lr_cv  = GridSearchCV(pipe_lr,  grid_lr,  cv=10).fit(X_train, Y_train)
svm_cv = GridSearchCV(pipe_svm, grid_svm, cv=10).fit(X_train, Y_train)
knn_cv = GridSearchCV(pipe_knn, grid_knn, cv=10).fit(X_train, Y_train)
dt_cv  = GridSearchCV(pipe_dt,  grid_dt,  cv=10).fit(X_train, Y_train)

# Recompute test accuracies
print("LR  test:", lr_cv.score(X_test, Y_test))
print("SVM test:", svm_cv.score(X_test, Y_test))
print("KNN test:", knn_cv.score(X_test, Y_test))
print("DT  test:", dt_cv.score(X_test, Y_test))
```

```python
# --- Visualize Classification Model Accuracies ---
import matplotlib.pyplot as plt
import numpy as np

# Collect cross-validation (CV) and test accuracies
cv_scores = {
    "Logistic Regression": lr_cv.best_score_,
    "SVM": svm_cv.best_score_,
    "Decision Tree": dt_cv.best_score_,
    "KNN": knn_cv.best_score_
}

test_scores = {
    "Logistic Regression": lr_cv.score(X_test, Y_test),
    "SVM": svm_cv.score(X_test, Y_test),
    "Decision Tree": dt_cv.score(X_test, Y_test),
    "KNN": knn_cv.score(X_test, Y_test)
}

# Extract for plotting
labels = list(cv_scores.keys())
cv_vals = [cv_scores[m] for m in labels]
test_vals = [test_scores[m] for m in labels]

x = np.arange(len(labels))
width = 0.35

# Create bar chart
plt.figure(figsize=(10, 5))
bars1 = plt.bar(x - width/2, cv_vals, width, label='Cross-Validation Accuracy')
bars2 = plt.bar(x + width/2, test_vals, width, label='Test Accuracy')

# Formatting
plt.ylabel('Accuracy')
plt.ylim(0, 1.05)
plt.title('Comparison of Model Accuracies')
plt.xticks(x, labels, rotation=15)
plt.legend()

# Annotate bars
def annotate_bars(bars):
    for b in bars:
        height = b.get_height()
        plt.text(b.get_x() + b.get_width()/2, height + 0.02, f"{height*100:.1f}%", 
                 ha='center', va='bottom', fontsize=9)

annotate_bars(bars1)
annotate_bars(bars2)

plt.tight_layout()
plt.show()
```

```python
# --- Confusion Matrix for Decision Tree ---
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Predict on test data
y_pred_dt = dt_cv.predict(X_test)

# Compute confusion matrix
cm = confusion_matrix(Y_test, y_pred_dt)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Failure', 'Success'])
disp.plot(cmap='Blues', values_format='d')

plt.title("Confusion Matrix - Decision Tree Classifier")
plt.show()
```
